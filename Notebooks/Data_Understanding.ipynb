## Data Understanding 

### Overview of the Dataset
- The dataset inculudes customers banking information, including demographic details, account information, financial records, transaction activity, and customer status. It provides a comprehensive view of customer banking behavior and can be used to analyze financial patterns, customer engagement, and operational performance.
### Types of Information Available
- This dataset includes informtion related to:

    **Customer Demographics**

    **Account details**

    **Loan Information**

    **Transaction Activity**

  **Customer Status**

## Column Understanding 
**Customer ID** : A unique identifier assigned to each customer<br>
**Customer Name** : Name of the customer<br>
**Gender** : Customer's gender.<br>
**Age** : Defines the Age of the customer<br>
**Occupation** : Customers's profession or employment category<br>
**Annual Income** : Customer's annual income in Indian Rupees.<br>
**City** : City where the customer resides.<br>
**State** : State where the customer resides.<br>
**Account Type** : Type of bank account (Saving, Salary, Current). <br>
**Account Open Date** : Date when the bank account was opened.<br>
**Account Balance** : Current balance availabe in the customer's account.<br>
**Avg Monthly Balance** : Average Balance maintained during a month.<br>
**Account Status** : Current status of the account(Active, Dormant, Closed)<br>
**Loan Type** :  Category of load taken by the customer ( Home, Car, Personal, Education, None)<br>
**Loan Amount** : Total Loan Amount Sanctioned to the cutomer.<br>
**EMI** : Monthly installment payable for the load.<br>
**Credit Score** : Numerical Representation of the customer's creditworthiness.<br>
**Monthly Transactions** : Total Number of transactions performed in a month.<br>
**Avg Transaction Amount** : Average value of each transaction.<br>
**UPI Txns** : Number of UPI transactions made in a month.<br>
**Last Transaction Date** : Date of the customer's most recent transaction.<br>
**Churn Status** : Indicates whether the customer has churned(Yes/No)

```python
# Unique values present in the dataset for further analysis (if Analysis Required)
def check_unique_values(df):
    unique_values = {
        "Gender": df['Gender'].unique(),
        "Account Type": df['Account Type'].unique(),
        "Account Status": df['Account Status'].unique(),
        "Loan Type" : df['Loan Type'].unique(),
        "Churn Status": df['Churn Status'].unique()
    
    }
    return unique_values

unique_values = check_unique_values(df)



for key, value in unique_values.items():
    print(f"{key} : {value}")
    


```

```python
# Range Validation
columns = [
    "AGE",
    'Annual Income (₹)',
    'Account Balance',
    'Avg Monthly Balance',
    'Loan Amount',
    'EMI',
    'Credit Score',
    'Monthly Transactions',
    'Avg Transaction Amount',
    'UPI Txns'
]

range_validation = pd.DataFrame({
    "Minimum Value" : df[columns].min(),
    "Maximum Value" : df[columns].max()
})
range_validation


# Expected Values of the Bussiness  Problems
expected_range = {
    'AGE' : '18-100',
    'Annual Income (₹)' : '>=10000 ',
    'Account Balance' : '> 0',
    'Avg Monthly Balance' : '>= 10000',
    'Loan Amount' : '>= 5000',
    'EMI' : '>0',
    'Credit Score': '200-900',
    'Monthly Transactions' : '>500',
    'Avg Transaction Amount' : '>0',
    'UPI Txns' : '>0'
}

range_validation['expected_range'] = range_validation.index.map(expected_range)
range_validation
```

```python
# Identify Quality Issue 
def data_quality_check(df):
    quality_report = {
        "total_records" : len(df),
        "duplicate_records" : df.duplicated().sum(),
        "missing_values" : df.isna().sum().to_dict(),
        "incorrect_age_value" : df[(df['AGE'] < 0) | (df['AGE'] > 100)].shape[0],
        "gender_description" : df['Gender'].value_counts().to_dict(),
        "incorrect_account_balance" : df[(df['Account Balance'] < 0)].shape[0],
    
    }
    return quality_report
quality_report = data_quality_check(df)


for key,value in quality_report.items():
    print(f"{key} : {value}")

```
